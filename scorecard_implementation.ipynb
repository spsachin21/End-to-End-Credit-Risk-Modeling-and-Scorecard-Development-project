# üìù Business Understanding

> You work for a financial institution that provides loans to urban customers. When evaluating loan applications, the institution faces two types of risk:

> - Rejecting a reliable applicant leads to a loss of potential business.
> - Approving a risky applicant may result in financial loss if the borrower defaults.

> The dataset contains historical loan application data, including the repayment status. The aim is to identify patterns that predict loan default, enabling smarter decisions like loan denial, reduced loan amounts, or higher interest rates for riskier profiles.

# üéØ Business Objectives
> - Defaulted loans represent the largest source of credit loss for lenders.
> - The objective is to identify high-risk applicants before approval using EDA and machine learning.
> - Understanding the key variables that indicate default risk helps improve lending decisions and manage the overall risk portfolio.

```python
def feature_selection_iv_summary(X_train, y_train):
    X_train_cat = X_train.select_dtypes(include='object').copy()
    X_train_num = X_train.select_dtypes(include='number').copy()

    X_train_num.fillna(X_train_num.mean(), inplace=True)

    chi2_check = {'Feature': [], 'Test Type': [], 'p-value': []}
    for column in X_train_cat:
        chi, p, dof, ex = chi2_contingency(pd.crosstab(y_train, X_train_cat[column]))
        chi2_check['Feature'].append(column)
        chi2_check['Test Type'].append("Chi-squared")
        chi2_check['p-value'].append(round(p, 10))

    chi2_result = pd.DataFrame(data=chi2_check)

    F_statistic, p_values = f_classif(X_train_num, y_train)
    anova_result = pd.DataFrame(data={
        'Feature': X_train_num.columns.values,
        'Test Type': 'ANOVA F',
        'p-value': p_values.round(decimals=10)
    })

    feature_stats = pd.concat([chi2_result, anova_result], ignore_index=True)

    feature_stats.sort_values(by=['Test Type', 'p-value'], ascending=[True, True], ignore_index=True, inplace=True)

    return feature_stats


feature_stats = feature_selection_iv_summary(X_train, y_train)
feature_stats
```

<!-- **Multicollinearity** -->

```python
def cramers_v(x, y):
    """Calculate Cram√©r's V for categorical-categorical associations."""
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

def correlation_ratio(categories, measurements):
    """Calculate the correlation ratio (eta squared) for categorical-numerical associations."""
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat) + 1
    y_avg = np.mean(measurements)
    numerator = np.sum([len(measurements[fcat == i]) * (np.mean(measurements[fcat == i]) - y_avg) ** 2 for i in range(cat_num)])
    denominator = np.sum((measurements - y_avg) ** 2)
    return np.sqrt(numerator / denominator)

def multicollinearity_matrix(X_train):
    num_features = X_train.select_dtypes(include=[np.number]).columns
    cat_features = X_train.select_dtypes(include=['object', 'category']).columns

    corr_matrix = pd.DataFrame(index=X_train.columns, columns=X_train.columns)

    for col1 in X_train.columns:
        for col2 in X_train.columns:
            if col1 == col2:
                corr_matrix.loc[col1, col2] = 1.0
            elif not pd.isna(corr_matrix.loc[col2, col1]):
                corr_matrix.loc[col1, col2] = corr_matrix.loc[col2, col1]
            else:
                if col1 in num_features and col2 in num_features:
                    corr = X_train[col1].corr(X_train[col2])

                elif col1 in cat_features and col2 in cat_features:
                    corr = cramers_v(X_train[col1], X_train[col2])

                elif col1 in cat_features and col2 in num_features:
                    corr = correlation_ratio(X_train[col1], X_train[col2])
                
                elif col1 in num_features and col2 in cat_features:
                    corr = correlation_ratio(X_train[col2], X_train[col1])
                
                corr_matrix.loc[col1, col2] = round(corr, 4)

    corr_matrix = corr_matrix.apply(pd.to_numeric)

    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Multicollinearity Matrix (Numerical and Categorical)')
    plt.show()

    return corr_matrix

correlation_matrix = multicollinearity_matrix(X_train)
correlation_matrix
```

```python
def identify_high_multicollinearity(corr_matrix, threshold=0.8):
    to_drop = set()
    already_checked = set()

    for col in corr_matrix.columns:
        for row in corr_matrix.index:
            if col != row and abs(corr_matrix.loc[row, col]) > threshold:
                if row not in already_checked:
                    to_drop.add(col)
                    already_checked.add(col)
    
    return list(to_drop)

features_to_drop = identify_high_multicollinearity(correlation_matrix, threshold=0.8)
features_to_drop
```

```python
import pandas as pd
import numpy as np

def calc_chi2(good1, bad1, good2, bad2):
    t1 = good1 + bad1
    t2 = good2 + bad2
    total = t1 + t2
    good_t = good1 + good2
    bad_t = bad1 + bad2
    if total == 0 or good_t == 0 or bad_t == 0:
        return 0
    if good1 == 0 or bad1 == 0 or good2 == 0 or bad2 == 0:
        return 0
    e_good1 = t1 * good_t / total
    e_bad1 = t1 * bad_t / total
    e_good2 = t2 * good_t / total
    e_bad2 = t2 * bad_t / total
    chi = ((good1 - e_good1)**2 / e_good1 +
           (bad1 - e_bad1)**2 / e_bad1 +
           (good2 - e_good2)**2 / e_good2 +
           (bad2 - e_bad2)**2 / e_bad2)
    return chi

class ManualCombiner:
    def __init__(self, max_initial_bins=500):
        self.rules = {}
        self.max_initial_bins = max_initial_bins

    def fit(self, X, y, min_samples_fraction=0.005):
        for col in X.columns:
            if pd.api.types.is_numeric_dtype(X[col]):
                self.rules[col] = self._chi_merge_numeric(X[col], y, min_samples_fraction)
            else:
                self.rules[col] = self._chi_merge_categorical(X[col], y, min_samples_fraction)

    def _chi_merge_numeric(self, x, y, min_samples_fraction):
        n = len(x)
        min_samples = int(min_samples_fraction * n)
        df = pd.DataFrame({'x': x, 'y': y}).dropna().sort_values('x')
        unique_count = df['x'].nunique()

        if unique_count > self.max_initial_bins:
            _, initial_splits = pd.qcut(df['x'], q=self.max_initial_bins, duplicates='drop', retbins=True)
            binned = pd.cut(df['x'], bins=initial_splits, include_lowest=True)
            freq = df.groupby(binned, observed=True).agg(
                total=('y', 'count'),
                bad=('y', 'sum')
            )
            uppers = list(initial_splits[1:])
        else:
            freq = df.groupby('x', sort=False).agg(
                total=('y', 'count'),
                bad=('y', 'sum')
            )
            uppers = freq.index.tolist()

        freq['good'] = freq['total'] - freq['bad']
        good_counts = freq['good'].values
        bad_counts = freq['bad'].values
        totals = freq['total'].values

        while True:
            num_bins = len(totals)
            if all(t >= min_samples for t in totals):
                break
            if num_bins <= 1:
                break

            good1 = good_counts[:-1]
            bad1 = bad_counts[:-1]
            good2 = good_counts[1:]
            bad2 = bad_counts[1:]

            chi2s = np.array([calc_chi2(g1, b1, g2, b2) for g1, b1, g2, b2 in zip(good1, bad1, good2, bad2)])

            min_idx = np.argmin(chi2s)
            good_counts[min_idx] += good_counts[min_idx + 1]
            bad_counts[min_idx] += bad_counts[min_idx + 1]
            totals[min_idx] += totals[min_idx + 1]
            uppers[min_idx] = uppers[min_idx + 1]  # Keep the upper bound of the merged bin

            good_counts = np.delete(good_counts, min_idx + 1)
            bad_counts = np.delete(bad_counts, min_idx + 1)
            totals = np.delete(totals, min_idx + 1)
            uppers = np.delete(uppers, min_idx)

        return [-np.inf] + uppers + [np.inf]

    def _chi_merge_categorical(self, x, y, min_samples_fraction):
        n = len(x)
        min_samples = int(min_samples_fraction * n)
        df = pd.DataFrame({'x': x.astype(str), 'y': y}).dropna()
        freq = df.groupby('x')['y'].agg(
            total='count',
            bad='sum'
        )
        freq['good'] = freq['total'] - freq['bad']
        freq['bad_rate'] = freq['bad'] / freq['total']
        freq.sort_values('bad_rate', inplace=True)
        categories = freq.index.tolist()
        good_counts = freq['good'].tolist()
        bad_counts = freq['bad'].tolist()
        totals = freq['total'].tolist()
        groups = [[cat] for cat in categories]

        while True:
            num_bins = len(totals)
            if all(t >= min_samples for t in totals):
                break
            if num_bins <= 1:
                break

            chi2s = []
            for i in range(num_bins - 1):
                chi = calc_chi2(good_counts[i], bad_counts[i], good_counts[i + 1], bad_counts[i + 1])
                chi2s.append(chi)

            min_idx = np.argmin(chi2s)
            good_counts[min_idx] += good_counts[min_idx + 1]
            bad_counts[min_idx] += bad_counts[min_idx + 1]
            totals[min_idx] += totals[min_idx + 1]
            groups[min_idx] += groups[min_idx + 1]
            del groups[min_idx + 1]
            del good_counts[min_idx + 1]
            del bad_counts[min_idx + 1]
            del totals[min_idx + 1]

        return {cat: label for label, grp in enumerate(groups) for cat in grp}

    def transform(self, X):
        X_binned = pd.DataFrame(index=X.index)
        for col in X.columns:
            if col in self.rules:
                rule = self.rules[col]
                if isinstance(rule, list):  # numeric
                    X_binned[col] = pd.cut(X[col], bins=rule, labels=range(len(rule) - 1), include_lowest=True).cat.codes
                elif isinstance(rule, dict):  # categorical
                    X_binned[col] = X[col].astype(str).map(rule).fillna(-1)
        return X_binned

class ManualWOETransformer:
    def __init__(self):
        self.woe_rules = {}

    def fit(self, X_binned, y):
        total_bad = y.sum()
        total_good = len(y) - total_bad
        for col in X_binned.columns:
            df = pd.DataFrame({'bin': X_binned[col], 'y': y})
            grouped = df.groupby('bin')['y'].agg(
                bad='sum',
                total='count'
            )
            grouped['good'] = grouped['total'] - grouped['bad']
            num_bins = grouped.shape[0]
            grouped['woe'] = np.log(
                ((grouped['bad'] + 0.5) / (total_bad + 0.5 * num_bins)) /
                ((grouped['good'] + 0.5) / (total_good + 0.5 * num_bins))
            )
            self.woe_rules[col] = grouped['woe'].to_dict()

    def transform(self, X_binned):
        X_woe = pd.DataFrame(index=X_binned.index)
        for col in X_binned.columns:
            if col in self.woe_rules:
                X_woe[col] = X_binned[col].map(self.woe_rules[col])
        return X_woe

    def fit_transform(self, X_binned, y):
        self.fit(X_binned, y)
        return self.transform(X_binned)

# Usage
combiner = ManualCombiner(max_initial_bins=500)  # Adjust if needed
combiner.fit(X_train, y_train, min_samples_fraction=0.005)

X_train_binned = combiner.transform(X_train)
X_test_binned = combiner.transform(X_test)

t = ManualWOETransformer()
X_train = t.fit_transform(X_train_binned, y_train)
X_test = t.transform(X_test_binned)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

class ManualScoreCard:
    def __init__(self,
                 combiner,               # your ManualCombiner instance
                 transer,                # your ManualWOETransformer instance
                 base_score=800,
                 base_odds=35,
                 pdo=20,
                 rate=2,
                 class_weight='balanced',
                 C=0.1,
                 **lr_kwargs):
        self.combiner = combiner
        self.transer  = transer
        
        self.lr = LogisticRegression(
            class_weight=class_weight,
            C=C,
            solver='liblinear',
            penalty='l1',
            random_state=42,
            **lr_kwargs
        )
        
        self.base_score = base_score
        self.base_odds  = base_odds
        self.pdo        = pdo
        self.rate       = rate
        
        self.coef_      = None
        self.intercept_ = None
        self.feature_names_ = None
        self.factor     = None
        self.offset     = None
        self.constant_per_var = None
        self.bin_mapping = {}   # {feature: {bin_code: (bin_label, woe, points)} }

    def fit(self, X_woe, y):
        self.feature_names_ = X_woe.columns.tolist()
        
        self.lr.fit(X_woe, y)
        self.coef_      = self.lr.coef_[0]
        self.intercept_ = self.lr.intercept_[0]
        
        self.factor  = self.pdo / np.log(self.rate)
        self.offset  = self.base_score - self.factor * np.log(self.base_odds)
        
        num_features = len(self.feature_names_)
        self.constant_per_var = (self.offset - self.factor * self.intercept_) / num_features if num_features > 0 else 0
        
        self._build_bin_mapping()
        
        return self

    def _build_bin_mapping(self):
        for i, col in enumerate(self.feature_names_):
            beta = self.coef_[i]
            
            rule = self.combiner.rules.get(col)
            if rule is None:
                continue
                
            woe_dict = self.transer.woe_rules.get(col, {})
            
            if isinstance(rule, list):  # numeric
                bins = rule
                bin_codes = sorted(woe_dict.keys())
                
                for idx, bin_code in enumerate(bin_codes):
                    right = bins[bin_code + 1]
                    right_str = 'inf' if np.isinf(right) else f"{right:g}"
                    
                    if bin_code == bin_codes[0]:
                        left_str = '-inf'
                    else:
                        left = bins[bin_code]
                        left_str = f"{left:g}"
                    
                    label = f"[{left_str} ~ {right_str})"
                    
                    woe = woe_dict.get(bin_code, np.nan)
                    points = self._woe_to_points(woe, beta)
                    self.bin_mapping.setdefault(col, {})[bin_code] = (label, woe, points)
                    
            elif isinstance(rule, dict):  # categorical
                group_to_bins = {}
                for cat, bin_code in rule.items():
                    group_to_bins.setdefault(bin_code, []).append(cat)
                
                for bin_code, cats in sorted(group_to_bins.items()):
                    lab = ', '.join(map(str, sorted(cats)))
                    woe = woe_dict.get(bin_code, np.nan)
                    points = self._woe_to_points(woe, beta)
                    self.bin_mapping.setdefault(col, {})[bin_code] = (lab, woe, points)

    def _woe_to_points(self, woe, beta):
        if np.isnan(woe) or beta == 0:
            return 0.0
        log_odds_contrib = - self.factor * beta * woe
        points = log_odds_contrib + self.constant_per_var
        return round(points, 2)

    def predict(self, X):
        X_binned = self.combiner.transform(X)
        X_woe    = self.transer.transform(X_binned)
        
        prob_bad = self.lr.predict_proba(X_woe)[:, 1]
        prob_good = 1 - prob_bad
        odds_good = prob_good / (prob_bad + 1e-10)  # avoid div0
        log_odds_good = np.log(odds_good)
        
        scores = self.offset + self.factor * log_odds_good
        return pd.Series(scores, index=X.index, name='score')

    def export(self, to_frame=True):
        rows = []
        for col in self.feature_names_:
            if col not in self.bin_mapping:
                continue
            for bin_code, (label, woe, points) in sorted(self.bin_mapping[col].items()):
                rows.append({
                    'name': col,
                    'value': label,
                    'score': points
                })
        
        df_card = pd.DataFrame(rows)
        
        if to_frame:
            return df_card
        else:
            return df_card.to_dict(orient='records')

# Usage
card = ManualScoreCard(
    combiner    = combiner,
    transer     = t,
    class_weight= 'balanced',
    C           = 0.1,
    base_score  = 800,
    base_odds   = 35,
    pdo         = 20,
    rate        = 2
)

card.fit(X_train, y_train)

final_card = card.export(to_frame=True)
final_card

# # To predict on raw test data (assuming X_test is raw features)
# test.loc[:, 'CreditScore'] = card.predict(X_test)
# # If X_test is already WOE-transformed, you may need to use raw features instead.
# # If needed, add a method for predicting from WOE:
# def predict_from_woe(self, X_woe):
#     prob_bad = self.lr.predict_proba(X_woe)[:, 1]
#     ...
#     return scores
```

```python
def print_score(y_true, y_pred, train=True):
    """
    y_pred can be:
    - class labels (0/1)
    - probabilities (preferred)
    """

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    prefix = "Train" if train else "Test"

    # Accuracy (only meaningful if y_pred are labels)
    if set(np.unique(y_pred)).issubset({0, 1}):
        acc = accuracy_score(y_true, y_pred)
        print(f"{prefix} Accuracy: {acc:.4f}")

    # AUC & KS (require probabilities)
    if not set(np.unique(y_pred)).issubset({0, 1}):
        auc = roc_auc_score(y_true, y_pred)
        fpr, tpr, _ = roc_curve(y_true, y_pred)
        ks = np.max(tpr - fpr)
        gini = 2 * auc - 1

        print(f"{prefix} AUC: {auc:.4f}")
        print(f"{prefix} KS:  {ks:.4f}")
        print(f"{prefix} Gini:{gini:.4f}")

    print("-" * 35)
```

```python
import numpy as np
import pandas as pd

def calculate_psi(expected, actual, buckets=10, bucket_type='percentiles', epsilon=1e-10):
    """
    Calculate Population Stability Index (PSI) between two distributions.
    
    Parameters:
    - expected: array-like (reference / training data)
    - actual: array-like (new / test / OOT data)
    - buckets: number of bins (default 10)
    - bucket_type: 'percentiles' (quantiles) or 'equal_width'
    - epsilon: small value to avoid log(0) or division by zero
    
    Returns: PSI value (float)
    """
    expected = np.asarray(expected)
    actual = np.asarray(actual)
    
    # Remove NaNs if any (or handle separately)
    expected = expected[~np.isnan(expected)]
    actual = actual[~np.isnan(actual)]
    
    if len(expected) == 0 or len(actual) == 0:
        return np.nan
    
    # Create bin edges based on reference (expected) data
    if bucket_type == 'percentiles':
        breakpoints = np.percentile(expected, np.linspace(0, 100, buckets + 1))
    else:  # equal width
        breakpoints = np.linspace(expected.min(), expected.max(), buckets + 1)
    
    # Avoid duplicate edges
    breakpoints = np.unique(breakpoints)
    
    # Calculate proportions
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _   = np.histogram(actual, bins=breakpoints)
    
    expected_prop = expected_counts / len(expected) + epsilon
    actual_prop   = actual_counts / len(actual) + epsilon
    
    # PSI per bin
    psi_per_bin = (actual_prop - expected_prop) * np.log(actual_prop / expected_prop)
    
    # Total PSI
    psi_total = np.sum(psi_per_bin)
    
    return psi_total
```
